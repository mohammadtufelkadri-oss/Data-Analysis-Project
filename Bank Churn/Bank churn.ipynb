# Missing Value Handling

## 11. Fill missing values in gender using mode.

```python
df['gender'].isnull().sum()
```

```python
df['gender']=df['gender'].fillna(df['gender'].mode()[0])
```

```python
df['gender'].isnull().sum()
```

## 12. Fill missing values in marital_status using mode

```python
df['marital_status'].isnull().sum()
```

```python
df['marital_status']=df.groupby('gender')['marital_status'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else np.nan)
```

```python
df['marital_status'].isnull().sum()
```

## 13. Fill missing values in annual_income using median.

```python
df['annual_income'].isnull().sum()
```

```python
df['annual_income']=df.groupby('account_type')['annual_income'].transform(lambda x: x.fillna(x.median()))
```

```python
df['annual_income'].isnull().sum()
```

## 14. Fill missing values in balance using median.

```python
df['balance'].isnull().sum()
```
